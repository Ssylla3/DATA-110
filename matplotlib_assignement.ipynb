```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Print available styles
print(plt.style.available)

# Set default style to 'seaborn-whitegrid'
plt.style.use('seaborn-v0_8-whitegrid')

# Choose another style for later
alt_style = 'ggplot'  # Example choice
plt.style.use(alt_style)  # Use alternative style

# Define the quadratic function
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1

```

1. Basic Line Plot

```python
# Define the quadratic function
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1

# Plot the function
plt.figure(figsize=(6,4))
plt.plot(x, y, label=r'$f(x) = x^2 - 2x + 1$', color='b')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Quadratic Function Plot")
plt.legend()
plt.grid(True)
plt.show()
```

2. Multiple Lines

```python
x = np.linspace(0.1, 3, 100)
f_x = x**3
g_x = np.exp(x)
h_x = np.log(x)

plt.figure(figsize=(6,4))
plt.plot(x, f_x, label=r'$x^3$', color='r')
plt.plot(x, g_x, label=r'$e^x$', color='g')
plt.plot(x, h_x, label=r'$\log(x)$', color='b')
plt.xlabel("x")
plt.ylabel("Function Value")
plt.title("Multiple Functions Plot")
plt.legend()
plt.show()
```

3.Styling Lines

```python
x = np.linspace(-2, 2, 100)
plt.figure(figsize=(6,4))
plt.plot(x, x**2, 'r-', label=r'$x^2$')  # Red solid line
plt.plot(x, x**3, 'g--', label=r'$x^3$') # Green dashed line
plt.plot(x, x**4, 'b-.', label=r'$x^4$') # Blue dash-dot line
plt.plot(x, x**5, 'm:', label=r'$x^5$')  # Magenta dotted line
plt.xlabel("x")
plt.ylabel("Function Value")
plt.title("Different Line Styles and Colors")
plt.legend()
plt.show()
```

# 4.Axes Limits and Scaling

```python
x = np.linspace(-10, 10, 1000)
y = 1 / (x**2 + 1)

plt.figure(figsize=(6,4))
plt.plot(x, y, label=r'$f(x) = \frac{1}{x^2+1}$')
plt.xlim(-10, 10)
plt.ylim(0, 1)
plt.legend()
plt.title("Custom Axes Limits")
plt.show()

# Reversed axes
plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.title("Reversed Axes")
plt.show()
```

5. Labels and legends

```python
x = np.linspace(0, 4, 100)
f_x = x
g_x = x**(1/3)

plt.figure(figsize=(6,4))
plt.plot(x, f_x, label=r'$f(x) = x$', color='r')
plt.plot(x, g_x, label=r'$g(x) = x^{1/3}$', color='b')
plt.xlabel("x")
plt.ylabel("Function Value")
plt.title("Labels and Legends Example")
plt.legend()
plt.show()
```

6. Basic scatter plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(6,4))
plt.scatter(x, y, color='b', label="Random Points")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Basic Scatter Plot")
plt.legend()
plt.show()
```

7. Customizing Scatter Plots

```python
x = np.linspace(0, 4*np.pi, 50)
y = x * np.sin(x)

plt.figure(figsize=(6,4))
plt.scatter(x, y, s=80, c='red', edgecolors='black', linewidth=1.5, label="Data Points")
plt.plot(x, y, 'g--', linewidth=2, label="Line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Customized Scatter Plot")
plt.legend()
plt.show()
```

8. combined plot

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = x * np.cos(x)

plt.figure(figsize=(6,4))
plt.plot(x, y, 'b-', label=r'$x\cos(x)$')
scatter_x = np.arange(-2*np.pi, 2*np.pi, np.pi/4)
scatter_y = scatter_x * np.cos(scatter_x)
plt.scatter(scatter_x, scatter_y, color='r', label="Sample Points")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Combined Line and Scatter Plot")
plt.legend()
plt.show()
```

9.Custom PLot with alternative style

```python
plt.style.use(alt_style)  # Use alternative style

x = np.linspace(-2*np.pi, 2*np.pi, 100)
tan_x = np.tan(x)
exp_x = x * np.exp(-x**2)
sin_x = np.sin(x) / x
arctan_x = np.arctan(x)

plt.figure(figsize=(8,5))
plt.plot(x, tan_x, 'r-', label=r'$tan(x)$')
plt.plot(x, exp_x, 'g--', label=r'$xe^{-x^2}$')
plt.plot(x, sin_x, 'b-.', label=r'$\frac{\sin(x)}{x}$')
plt.plot(x, arctan_x, 'm:', label=r'$arctan(x)$')

plt.ylim(-5, 5)  # Limit y to avoid extreme values
plt.xlabel("x")
plt.ylabel("Function Value")
plt.title("Custom Plot with Alternative Style")
plt.legend()
plt.grid(True)
plt.show()
```

10. Comparison

```python
# Define x values
x = np.linspace(-2, 2, 100)

# Define functions
y1 = x**2
y2 = x**3
y3 = x**4
y4 = x**5

# Plot using 'seaborn-whitegrid' style
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(6,4))
plt.plot(x, y1, 'r-', label=r'$x^2$')  # Red solid line
plt.plot(x, y2, 'g--', label=r'$x^3$') # Green dashed line
plt.plot(x, y3, 'b-.', label=r'$x^4$') # Blue dash-dot line
plt.plot(x, y4, 'm:', label=r'$x^5$')  # Magenta dotted line
plt.xlabel("x")
plt.ylabel("Function Value")
plt.title("Line Styles and Colors - Seaborn-Whitegrid Style")
plt.legend()
plt.grid(True)
plt.show()

# Plot using 'ggplot' style
plt.style.use('ggplot')
plt.figure(figsize=(6,4))
plt.plot(x, y1, 'r-', label=r'$x^2$')  # Red solid line
plt.plot(x, y2, 'g--', label=r'$x^3$') # Green dashed line
plt.plot(x, y3, 'b-.', label=r'$x^4$') # Blue dash-dot line
plt.plot(x, y4, 'm:', label=r'$x^5$')  # Magenta dotted line
plt.xlabel("x")
plt.ylabel("Function Value")
plt.title("Line Styles and Colors - GGPlot Style")
plt.legend()
plt.grid(True)
plt.show()
```

1 Seaborn-Whitegrid Style

Advantages

✔ Clean and Minimalistic – The white background with subtle gridlines makes the plot easy to read.
✔ Professional Look – Ideal for presentations, academic papers, and reports where clarity is key.
✔ Easy Distinction Between Lines – The white background allows the different line colors to pop clearly.

Disadvantages

✖ Less Contrast – The grid is light and might not stand out well on some screens.
✖ Lacks Strong Visual Impact – While neat, it does not add much emphasis to the data beyond simple readability.

2️ GGPlot Style

 Advantages

✔ High Contrast – The gray background with prominent gridlines makes the graph pop.
✔ Better for Data Exploration – Inspired by R’s ggplot2, it is excellent for spotting trends in large datasets.
✔ Bold and Eye-catching – The stronger gridlines and background make the graph stand out.

Disadvantages

✖ Can Feel Too Busy – The darker background might make the plot look cluttered.
✖ Not Ideal for Certain Uses – If printed in black & white, the contrast might not be as effective.


Conclusion
Use Seaborn-Whitegrid if you want a clean and professional look, especially for reports or formal presentations.
Use GGPlot if you need high contrast and bold visualization, making it great for exploratory data analysis or visual storytelling.
