```python
# Install & import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#  Set style for plots
sns.set(style="whitegrid")

```

```python
import pandas as pd

# Load directly
df = pd.read_csv("us_gun_deaths.csv")
df.head()
```

```python
# Clean the Data
# Convert victim_age to numeric
df['victim_age'] = pd.to_numeric(df['victim_age'], errors='coerce')

# Quick data check
print(df.info())
print(df.head())

```

```python

# ========================
# ðŸ“Š VISUALIZATION 1: Gun Deaths by Year
# ========================
deaths_per_year = df['year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x=deaths_per_year.index, y=deaths_per_year.values, marker='o')
plt.title("Gun Deaths in the U.S. by Year")
plt.xlabel("Year")
plt.ylabel("Number of Deaths")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Graph 1: Gun Deaths in the U.S. by Year (Line Chart)
1. What it shows:
This line chart illustrates the number of gun-related deaths reported each year from 1985 to 2023.

2. Deeper Analysis:
The line is relatively stable with high values each year, indicating that gun deaths are not an isolated or recent issueâ€”theyâ€™ve been persistent for nearly four decades.

There may be slight fluctuations, but no clear long-term downward trend. If anything, the number appears to plateau at a high level.

This challenges the notion that gun violence is only a modern or post-social media phenomenonâ€”it's deeply rooted and consistent over time.

Policymakers and the public may perceive certain years as â€œworseâ€ based on media attention, but the data shows that the problem has always been significant.

This graph suggests that gun policy, prevention efforts, or social programs may not have made a substantial nationwide impact over this time span

```python
# ðŸ“Š VISUALIZATION 2: Victim Sex Distribution
# ========================
sex_counts = df['victim_sex'].value_counts()

plt.figure(figsize=(7, 5))
sns.barplot(x=sex_counts.index, y=sex_counts.values)
plt.title("Gun Deaths by Victim Sex")
plt.xlabel("Sex")
plt.ylabel("Number of Deaths")
plt.tight_layout()
plt.show()
```

Graph 2: Gun Deaths by Victim Sex (Bar Chart)
1. What it shows:
This bar chart compares the number of gun-related deaths between male, female, and unknown-sex victims.

2. Deeper Analysis:
Males make up the vast majority of gun-related deaths, indicating a strong gender disparity in victimization.

Possible explanations could include higher male involvement in risky environments, crime, or firearm ownership.

Females account for a much smaller portion, but that doesn't mean their experiences are less importantâ€”female deaths may correlate more with domestic violence, which is a distinct issue requiring separate policy attention.

The â€œUnknownâ€ category, though small, suggests data limitationsâ€”emphasizing the need for accurate and complete data collection in law enforcement and public health reporting.

This visualization could guide prevention strategies that are tailored differently for men and women, focusing on behavioral, community, or mental health factors.





```python
# ðŸ“Š VISUALIZATION 3: Gun Deaths by Region
# ========================
region_counts = df['region'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=region_counts.values, y=region_counts.index)
plt.title("Gun Deaths by Region")
plt.xlabel("Number of Deaths")
plt.ylabel("Region")
plt.tight_layout()
plt.show()
```

Graph 3: Gun Deaths by Region (Horizontal Bar Chart)
1. What it shows:
This horizontal bar chart breaks down the total number of deaths by U.S. region (e.g., Southeast, Southwest, Northeast, etc.).

2. Deeper Analysis:
The Southeast and Southwest regions have significantly higher death counts than others.

This could reflect several factors: population density, access to firearms, poverty levels, cultural attitudes toward guns, or the strength (or lack) of local gun control laws.

The Northeast and West Coast, while still affected, have comparatively lower numbersâ€”potentially reflecting stronger regulations or urban-rural population differences.

This graph reinforces the idea that gun violence is not equally distributed across the countryâ€”itâ€™s concentrated in specific geographic regions.

These insights could support regionally-targeted interventions instead of â€œone-size-fits-allâ€ national campaigns.
