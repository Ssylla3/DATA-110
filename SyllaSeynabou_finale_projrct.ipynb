Title: Tracing the Footprints of Terror: Analyzing Global Terror Attacks (2018–2023)

1. Introduction

Terrorism is a global issue with devastating and lasting impacts. Through this project, I sought to uncover where and how terrorist attacks have occurred, who has been responsible, and what patterns can be observed over time. I selected a simplified dataset inspired by the Global Terrorism Database that includes six attacks from 2018 to 2023 across Afghanistan, Iraq, and Nigeria. This dataset includes location, group responsible, attack type, number of deaths and injuries, and the year. My goal was to turn this data into a story that makes the human impact of terrorism easier to understand through visualizations.

2. EDA Process & Key Findings

Using Python (Pandas) for data exploration, I examined the distribution of attacks across years, locations, groups, and casualties.

Key Findings:

- Deaths by Country: Afghanistan and Nigeria experienced the highest death tolls, with 410 each, followed by Iraq.

- Injuries by Group: The Taliban caused the most injuries (550), with Boko Haram and ISIS close behind.

- Yearly Trends: Deaths declined slightly until 2021, followed by a noticeable increase in 2022 and 2023.

- Attack Methods: All three groups used both bombings and armed assaults, showing tactical adaptability.

- Deadliest Cities: Kabul and Maiduguri were the most impacted cities, with deaths exceeding 200.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample terrorism dataset
data = {
    'year': [2018, 2019, 2020, 2021, 2022, 2023],
    'country': ['Afghanistan', 'Afghanistan', 'Iraq', 'Iraq', 'Nigeria', 'Nigeria'],
    'attack_type': ['Bombing', 'Armed Assault', 'Bombing', 'Armed Assault', 'Bombing', 'Armed Assault'],
    'deaths': [230, 180, 150, 120, 200, 210],
    'wounded': [300, 250, 200, 150, 180, 190],
    'group': ['Taliban', 'Taliban', 'ISIS', 'ISIS', 'Boko Haram', 'Boko Haram'],
    'city': ['Kabul', 'Kandahar', 'Baghdad', 'Mosul', 'Lagos', 'Maiduguri']
}

df = pd.DataFrame(data)
```

Visualization 1: Bar Chart

```python
# Aggregate deaths by country
deaths_by_country = df.groupby('country')['deaths'].sum().sort_values(ascending=False)

# Plot
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(deaths_by_country.index, deaths_by_country.values, color='pink')
ax.set_title('Deaths by Country (2018–2023)', fontsize=16, color='pink', pad=20)
ax.set_xlabel('Country', fontsize=12, color='white')
ax.set_ylabel('Total Deaths', fontsize=12, color='white')
ax.tick_params(colors='white')
plt.tight_layout()
plt.show()
```

 Deaths by Country

This static visualization highlights how terrorism has disproportionately affected certain regions. Afghanistan and Nigeria lead in total deaths, signaling the persistence of violence in conflict-heavy zones.

Visualization 2: Line Chart

```python
# Aggregate deaths per year
deaths_per_year = df.groupby('year')['deaths'].sum()

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(deaths_per_year.index, deaths_per_year.values, marker='o', color='pink')
ax.set_title('Total Deaths Over Time (2018–2023)', fontsize=16, color='pink', pad=20)
ax.set_xlabel('Year', fontsize=12, color='white')
ax.set_ylabel('Deaths', fontsize=12, color='white')
ax.tick_params(colors='white')
plt.style.use('dark_background')
plt.tight_layout()
plt.show()

```

Deaths Over Time

This time series view shows a dip in death tolls through 2021, followed by a sharp increase. It illustrates how terrorism can quickly resurge after periods of decline

Visualization 3:Stacked Bar Chart

```python
# Group by attack type and group
attack_group = df.groupby(['group', 'attack_type'])['deaths'].sum().unstack().fillna(0)

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
attack_group.plot(kind='bar', stacked=True, ax=ax, color=['lightpink', 'deeppink'])
ax.set_title('Attack Types by Terrorist Group (2018–2023)', fontsize=16, color='pink', pad=20)
ax.set_xlabel('Terrorist Group', fontsize=12, color='white')
ax.set_ylabel('Total Deaths', fontsize=12, color='white')
ax.tick_params(colors='white')
plt.legend(title='Attack Type', facecolor='black', edgecolor='white', labelcolor='white', title_fontsize=10, fontsize=9)
plt.style.use('dark_background')
plt.tight_layout()
plt.show()

```

Attack Types by Group

This chart reveals that no single group relied on one method. The Taliban, Boko Haram, and ISIS all used bombings and assaults, indicating flexible and evolving strategies.
