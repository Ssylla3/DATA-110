```python
# Task 1: Bar Chart for Crime Rates in 2023
import matplotlib.pyplot as plt

# Data
offenses = [
    "Aggravated Assault", "Burglary", "Larceny", "Motor Vehicle Theft",
    "Murder", "Rape", "Robbery"
]
crime_rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]

# Bar Chart for Crime Rates in 2023
plt.figure(figsize=(12, 6))
bars = plt.bar(offenses, crime_rates_2023, color='royalblue', alpha=0.75)
plt.xlabel("Crime Type", fontsize=12)
plt.ylabel("Offenses per 100,000 People (2023)", fontsize=12)
plt.title("Crime Rates per 100,000 People in 2023", fontsize=14, fontweight='bold')
plt.xticks(rotation=30, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add values on top of bars for better readability
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
             f'{bar.get_height():.1f}', ha='center', fontsize=10, fontweight='bold')

plt.show()
```

```python
# Task 2: Horizontal Bar Chart for Change in Crime Rates (2022-2023)
import matplotlib.pyplot as plt
import numpy as np

# Data
offenses = [
    "Aggravated Assault", "Burglary", "Larceny", "Motor Vehicle Theft",
    "Murder", "Rape", "Robbery"
]
change_rates = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

# Generate unique colors for each bar
colors = plt.cm.Paired(np.linspace(0, 1, len(offenses)))

# Version 1: Basic Horizontal Bar Chart
plt.style.use('default')  # Ensure white background
plt.figure(figsize=(12, 6))
plt.barh(offenses, change_rates, color=colors, alpha=0.85)
plt.xlabel("Change in Offenses per 100,000 People (2022-2023)", fontsize=12)
plt.ylabel("Crime Type", fontsize=12)
plt.title("Percentage Change in Crime Rates (2022-2023)", fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add values at the end of bars for readability
for index, value in enumerate(change_rates):
    plt.text(value + (0.5 if value > 0 else -3), index, f'{value:.1f}%',
             va='center', fontsize=10, fontweight='bold')

plt.show()

# Version 2: Improved Readability with Bolder Colors and Labels
plt.style.use('default')  # Ensure white background
plt.figure(figsize=(12, 6))
bars = plt.barh(offenses, change_rates, color=colors, alpha=0.85, edgecolor='black')
plt.xlabel("Change in Offenses per 100,000 People (2022-2023)", fontsize=12, fontweight='bold')
plt.ylabel("Crime Type", fontsize=12, fontweight='bold')
plt.title("Crime Rate Change from 2022 to 2023", fontsize=14, fontweight='bold', color='darkred')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7, linewidth=0.8)

# Adding text labels inside bars for better readability
for bar in bars:
    plt.text(bar.get_width() + (0.5 if bar.get_width() > 0 else -3),
             bar.get_y() + bar.get_height()/2, f'{bar.get_width():.1f}%',
             va='center', fontsize=10, fontweight='bold', color='black')

plt.show()

# Version 3: Dark Background for Higher Contrast
plt.style.use('dark_background')  # Apply dark background ONLY for version 3
plt.figure(figsize=(12, 6))
bars = plt.barh(offenses, change_rates, color=colors, alpha=0.85, edgecolor='white')
plt.xlabel("Change in Offenses per 100,000 People (2022-2023)", fontsize=12, fontweight='bold', color='white')
plt.ylabel("Crime Type", fontsize=12, fontweight='bold', color='white')
plt.title("Crime Rate Change from 2022 to 2023", fontsize=14, fontweight='bold', color='lightcoral')
plt.xticks(fontsize=10, color='white')
plt.yticks(fontsize=10, color='white')
plt.grid(axis='x', linestyle='--', alpha=0.5, linewidth=0.8, color='white')

# Adding text labels inside bars for better readability
for bar in bars:
    plt.text(bar.get_width() + (0.5 if bar.get_width() > 0 else -3),
             bar.get_y() + bar.get_height()/2, f'{bar.get_width():.1f}%',
             va='center', fontsize=10, fontweight='bold', color='white')

plt.show()

# Version 4: Improved Version with Black Percentage Text Inside Bars
plt.style.use('default')  # Reset to white background
plt.figure(figsize=(12, 6))
bars = plt.barh(offenses, change_rates, color=colors, alpha=0.85, edgecolor='black')

plt.xlabel("Change in Offenses per 100,000 People (2022-2023)", fontsize=12, fontweight='bold', color='black')
plt.ylabel("Crime Type", fontsize=12, fontweight='bold', color='black')
plt.title("Crime Rate Change from 2022 to 2023", fontsize=14, fontweight='bold', color='darkred')
plt.xticks(fontsize=10, color='black')
plt.yticks(fontsize=10, color='black')
plt.grid(axis='x', linestyle='--', alpha=0.7, linewidth=0.8, color='black')

# Adding text labels inside bars with black font
for bar in bars:
    text_x = bar.get_width() * 0.5  # Positioning text inside the bar
    text_y = bar.get_y() + bar.get_height()/2  # Centering text vertically

    plt.text(text_x, text_y, f'{bar.get_width():.1f}%',
             va='center', ha='center', fontsize=10, fontweight='bold', color='black')

plt.show()
```

Why This Bar Graph Makes More Sense Than Task 1's Bar Graph
1. Absolute Numbers Can Be Misleading

Task 1’s graph shows the total crime rates per 100,000 people for each offense in 2023. However, crimes like larceny (1,347.2 per 100,000 people) will always appear larger than crimes like murder (5.7 per 100,000 people) simply because certain crimes happen more frequently.
This can give the impression that some crimes are far more concerning than others, when in reality, the rate of change over time is a better measure of trends.
2. Percentage Changes Provide Context

Task 2’s graph focuses on the percentage change in crime rates from 2022 to 2023, which shows whether crime is increasing or decreasing.
For example:
Murder decreased by -12.0%, which is a significant drop.
Motor vehicle theft increased by +12.0%, indicating a rising concern.
This makes it clear which crimes are getting worse and which are improving.
3. Helps Identify Trends Over Time

If a crime had a high total number in Task 1 but decreased significantly in Task 2, it shows positive progress (e.g., burglary dropped by -8.1%).
If a crime had a low total number but increased, it could indicate emerging risks that need attention (e.g., motor vehicle theft increased despite lower absolute numbers).
Trends over time are more useful for policymakers, law enforcement, and communities than just knowing how common a crime was in one year.
4. Real-World Application

Law enforcement can allocate resources more efficiently based on rising or falling trends.
Policy changes can be made to address growing concerns (e.g., increase security measures to combat rising car theft).
Communities can use this data to understand which crimes are declining and where more prevention efforts are needed.


```python
 # Combined bar chart comparing crime rates in 2022 and 2023:
 import matplotlib.pyplot as plt
import numpy as np

# Data
offenses = [
    "Aggravated Assault", "Burglary", "Larceny", "Motor Vehicle Theft",
    "Murder", "Rape", "Robbery"
]
crime_rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
change_rates = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

# Calculate crime rates for 2022 based on 2023 values and percentage changes
crime_rates_2022 = [round(crime_rates_2023[i] / (1 + (change_rates[i] / 100)), 1) for i in range(len(offenses))]

# Set up the figure
plt.figure(figsize=(12, 6))
bar_width = 0.4  # Width of the bars
x_indexes = np.arange(len(offenses))  # Positions for bars

# Create the bars for 2022 and 2023
bars1 = plt.bar(x_indexes - bar_width/2, crime_rates_2022, width=bar_width, color='darkorange', label="2022")
bars2 = plt.bar(x_indexes + bar_width/2, crime_rates_2023, width=bar_width, color='royalblue', label="2023")

# Labels and title
plt.xlabel("Crime Type", fontsize=12)
plt.ylabel("Offenses per 100,000 People", fontsize=12)
plt.title("Comparison of Crime Rates in 2022 and 2023", fontsize=14, fontweight='bold')
plt.xticks(ticks=x_indexes, labels=offenses, rotation=30, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12)

# Add values on top of bars for better readability
for bars in [bars1, bars2]:
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
                 f'{bar.get_height():.1f}', ha='center', fontsize=10, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
